Step 1. Import the necessary libraries

```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt

users = pd.read_csv('user.txt', delimiter = "|")
users.head()




```

```python
# Step 4. Discover what is the mean age per occupation

meanAge = users.groupby('occupation')['age'].mean()
meanAge
```

```python
# Step 5. Discover the Male ratio per occupation and sort it from the most to the least

dfMale = users.groupby(['occupation', 'gender'])['user_id'].sum().rename("count")
dfMale / dfMale.groupby(level=0).sum() 
```

```python
# Step 6. For each occupation, calculate the minimum and maximum ages

df = users
df = (df.assign(age=df['age'])
       .groupby(['occupation'])['age'].agg([('Min' , 'min'), ('Max', 'max')])
       .add_prefix('Age'))

df
```

```python
# Step 7. For each combination of occupation and sex, calculate the mean age

meanAge2 = users.groupby(['occupation', 'gender'])['age'].mean()
meanAge2
```

```python
# Step 8. For each occupation present the percentage of women and men

df2 = users.groupby(['occupation', 'gender'])['user_id'].sum().rename("count")
df2 / df2.groupby(level=0).sum() *100
```

```python
import numpy as np
import pandas as pd
#import matplotlib.pyplot as plt

ser1 = np.random.randint(1,5, size=100)
ser2 = np.random.randint(1,4, size=100)
ser3 = np.random.randint(10000,30001, size=100)
#ser3
```

```python
# Step 3. Create a DataFrame by joinning the Series by column

dfH1 = pd.DataFrame([ser1]).transpose()
dfH2 = pd.DataFrame([ser2]).transpose()
dfH3 = pd.DataFrame([ser3]).transpose()
#dfH3

dfHouse = pd.concat([dfH1, dfH2, dfH3], axis=1)

dfHouse
```

```python
# Step 4. Change the name of the columns to bedrs, bathrs, price_sqr_meter

dfHouse.columns = ['bedrs','bathrs','price_sqr_meter']
dfHouse.head()
```

```python
# Step 5. Create a one column DataFrame with the values of the 3 Series and assign it to 'bigcolumn'

dfCombine['bigcolumn'] = dfHouse['bedrs'].astype(str) + dfHouse['bathrs'].astype(str) + dfHouse['price_sqr_meter'].astype(str)

dfCombine = dfCombine[['bigcolumn']]
dfCombine
```

```python
# Step 6. Ops it seems it is going only until index 99. Is it true?

#Answer: Yes; index starts at 0
```

```python
# Step 7. Reindex the DataFrame so it goes from 0 to 299

dfH7 = dfCombine
dfH7.reindex([0, 299])


dfH7
```

Question 6
Create a line plot showing the number of marriages and divorces per capita in the U.S. between 1867 and 2014. Label both lines and show the legend.
Don't forget to label your axes!

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

divData = pd.read_csv('us-marriages-divorces-1867-2014.csv')
divData.head()
```

```python
# Marriages and divorces per capita line plot

info1 = divData[["Year", "Marriages_per_1000", "Divorces_per_1000"]]
info2 = info1.melt('Year', var_name='cols',  value_name='Count')
sns.lineplot(x="Year", y="Count", hue='cols', data=info2)
plt.show()
```

Question 7

Create a vertical bar chart comparing the number of marriages and divorces per capita in the U.S. between 1900, 1950, and 2000.
Don't forget to label your axes!

```python

info3 = divData.loc[(divData['Year'] == 1900) | (divData['Year'] == 1950) | 
                 (divData['Year'] == 2000)] [["Year", "Marriages_per_1000", "Divorces_per_1000"]]

#info3
info4 = info3.melt('Year', var_name="Column", value_name="Count")
sns.catplot(x='Year', y='Count', hue='Column', data=info4, kind='bar')

plt.show()
```
